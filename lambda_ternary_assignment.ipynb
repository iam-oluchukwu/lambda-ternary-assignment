# Lambda with Ternary Operator (Map Function)
## Objective
This program uses lambda functions and a ternary operator within the map() function.

## Method
- Iterate through a list using map()
- Use lambda to check each element
- Square integers
- Return None for non-integer values

```python
data = [5, 'Python', 3.14, -2, 'Data', True, 0, 12.5, 9, 'AI', None, 4]

result = list(map(lambda x: x * x if type(x) == int else None, data))

print(result)
```
